```python
df.columns = ["key","date","offense","ageGroup","sex","race","borough","longitude_latitude"]
# key,date,offense,ageGroup,sex,race,borough,longitude_latitude,month,year
df
```

```python
df = df.dropna(subset=['borough'])
df = df.drop(df.index[-1]) # drop the header column
df["date"] = pd.to_datetime(df["date"]).dt.date
df["month"] = pd.DatetimeIndex(df["date"]).month
df["year"] = pd.DatetimeIndex(df["date"]).year.astype(float)
df
```

```python
df_1819 = df[df['year'].isin([2018,2019])]
df_1819
```

```python
df_1819['borough'] = df_1819['borough'].replace(['NEW YORK'],'MANHATTAN')
```

```python
print("Number of summon cases by borough in 2018, ranked from high to low")
print(df_1819[df_1819["year"]==2018].groupby(["borough"]).count().sort_values(by='date',ascending=False)['date'])
print("\nNumber of summon cases by borough in 2019 , ranked from high to low")
print(df_1819[df_1819["year"]==2019].groupby(["borough"]).count().sort_values(by='date',ascending=False)['date'])

print("\nNumber of summon cases by ageGroup count in 2018, ranked from high to low")
print(df[df["year"]==2018].groupby(["ageGroup"]).count().sort_values(by='date',ascending=False)['date'])
print("\nNumber of summon cases by ageGroup count in 2019, ranked from high to low")
print(df[df["year"]==2019].groupby(["ageGroup"]).count().sort_values(by='date',ascending=False)['date'])

print("\nNumber of summon cases by race in 2018, ranked from high to low")
print(df[df["year"]==2018].groupby(["race"]).count().sort_values(by='date',ascending=False)['date'])
print("\nNumber of summon cases by race in 2019, ranked from high to low")
print(df[df["year"]==2019].groupby(["race"]).count().sort_values(by='date',ascending=False)['date'])


print("\nNumber of summon cases by sex in 2018, ranked from high to low")
print(df[df["year"]==2018].groupby(["sex"]).count().sort_values(by='date',ascending=False)['date'])
print("\nNumber of summon cases by sex in 2019, ranked from high to low")
print(df[df["year"]==2019].groupby(["sex"]).count().sort_values(by='date',ascending=False)['date'])


```
